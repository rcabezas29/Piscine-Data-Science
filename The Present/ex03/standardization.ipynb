```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Train_knight.csv")
df
```

```python
for col in df:
	try:
		df[col] = (df[col] - df[col].mean()) / df[col].std()
	except:
		continue
df
```

```python
colors = {'Sith': 'red', 'Jedi': 'blue'}
color_list = [colors[group] for group in df['knight']]

fig, axs = plt.subplots(2)
fig.set_figwidth(3)
axs[0].scatter(df['Sensitivity'], df['Hability'], c=color_list, alpha=0.5)
axs[1].scatter(df['Strength'], df['Power'], c=color_list, alpha=0.5)
plt.show()
```
