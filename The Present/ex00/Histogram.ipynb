```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv("../Train_knight.csv")
df2
```

```python
num_features = len(df2.columns) - 1

for i, feature in enumerate(df2.columns[:-1]):
    plt.subplot(5, 6, i + 1)
    for knight_type, color in zip(['Sith', 'Jedi'], ['red', 'blue']):
        subset = df2[df2['knight'] == knight_type][feature]
        subset.hist(bins=20, alpha=0.5, label=f'{knight_type}', color=color, figsize=(15, 15))
        plt.title(feature)
        plt.legend()

plt.tight_layout()
plt.show()
```
