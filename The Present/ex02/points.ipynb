```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("../Train_knight.csv")
df2 = pd.read_csv("../Test_knight.csv")
df1
```

```python
colors = {'Sith': 'red', 'Jedi': 'blue'}
color_list = [colors[group] for group in df1['knight']]

fig, axs = plt.subplots(2, 2)
axs[0, 0].scatter(df1['Sensitivity'], df1['Hability'], c=color_list, alpha=0.5)
axs[0, 1].scatter(df1['Strength'], df1['Power'], c=color_list, alpha=0.5)
axs[1, 0].scatter(df2['Sensitivity'], df2['Hability'], c='green', alpha=0.5)
axs[1, 1].scatter(df2['Strength'], df2['Power'], c='green', alpha=0.5)
plt.show()
```
